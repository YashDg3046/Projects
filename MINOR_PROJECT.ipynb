```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
uploaded = files.upload()
```

```python
ds_tracks=pd.read_csv("spotify dataset.csv")
ds_tracks.head()
```

```python
pd.isnull(ds_tracks).sum()
```

```python
ds_tracks.info()
```

```python
sorted_ds= ds_tracks.sort_values('track_popularity',ascending =True).head(10)
sorted_ds
```

```python
ds_tracks.describe().transpose()
```

```python
most_popular = ds_tracks.query('track_popularity>90',inplace = False).sort_values('track_popularity', ascending = False)
most_popular[:10]
```

```python
ds_tracks.set_index("track_album_release_date ", inplace= True)
ds_tracks.index = pd.to_datetime(ds_tracks.index)
ds_tracks.head()
```

```python
ds_tracks[["track_artist"]].iloc[18]
```

```python
ds_tracks.duration.head()
```

```python
ds_tracks["duration"] = ds_tracks["track_duration"].apply(lambda x: round(x/4000))
ds_tracks.drop("track_duration", inplace=True, axis =1)
```

```python
corr_ds=ds_tracks.drop(['key', 'mode'],axis =1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_ds,annot=True,fmt=".1g", vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation Heatmap")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_ds = ds_tracks.sample(int(0.004*len(ds_tracks)))
print(len(sample_ds))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=ds_tracks, y="loudness",x="energy", color ="c").set(title = "Loudness Vs Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot (data = ds_tracks,y="track_popularity",x="acousticness",color ="b").set(title = "Popularity Vs Acousticness")
```

```python
ds_tracks['dates']=ds_tracks.index.get_level_values('track_album_release_date')
ds_tracks.dates=pd.to_datetime(ds_tracks.dates)
years=ds_tracks.dates.dt.year
```

```python
total_ds = ds_tracks.duration
sns.set_style(style="whitegrid")
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years, y= total_ds,ax=ax).set_title("Years Vs Duration")
plt.xticks(rotation=60)
```

```python
ds_genre=pd.read_csv("spotify dataset.csv")
ds_genre.head()
```

```python
plt.title("Duration of songs ")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y='playlist_genre', x='duration_ms',data=ds_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize=(10,5))
famous = ds_genre.sort_values("track_popularity", ascending=False).head(10)
sns.barplot(y='playlist_genre',x='track_popularity',data = famous).set_title("Top 5 Genre with high popularity")
```
